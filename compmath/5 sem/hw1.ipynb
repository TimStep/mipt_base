## 1

$$ ay^3 + d = 0 \implies y^3 = -\frac{d}{a} \implies y = -\sqrt[3]{\frac{d}{a}} $$

Обозначим $ y = f(a, d) = -\left( \frac{d}{a} \right)^{1/3} $.

$$
\Delta y \approx \left| \frac{\partial f}{\partial a} \right| \Delta a + \left| \frac{\partial f}{\partial d} \right| \Delta d
$$

Частные производные:
$$
\frac{\partial y}{\partial a} = -\frac{\partial}{\partial a} \left( \frac{d}{a} \right)^{1/3} = -\frac{1}{3} \left( \frac{d}{a} \right)^{-2/3} \cdot \left( -\frac{d}{a^2} \right) = \frac{1}{3} \left( \frac{d}{a} \right)^{-2/3} \cdot \frac{d}{a^2}
$$
$$
\frac{\partial y}{\partial d} = -\frac{\partial}{\partial d} \left( \frac{d}{a} \right)^{1/3} = -\frac{1}{3} \left( \frac{d}{a} \right)^{-2/3} \cdot \frac{1}{a}
$$

Подставим значения $ a = 1 $, $ d = 8 $:
$$
\frac{d}{a} = 8, \quad \left( \frac{d}{a} \right)^{-2/3} = 8^{-2/3} = \frac{1}{8^{2/3}} = \frac{1}{4}
$$

Тогда:
$$
\left| \frac{\partial y}{\partial a} \right| = \frac{1}{3} \cdot \frac{1}{4} \cdot \frac{8}{1^2} = \frac{1}{3} \cdot \frac{1}{4} \cdot 8 = \frac{2}{3}
$$
$$
\left| \frac{\partial y}{\partial d} \right| = \frac{1}{3} \cdot \frac{1}{4} \cdot \frac{1}{1} = \frac{1}{12}
$$

Таким образом:
$$
\begin{aligned}
\Delta y \approx \left| \frac{\partial y}{\partial a} \right| \Delta a + \left| \frac{\partial y}{\partial d} \right| \Delta d &= \frac{2}{3} \cdot 10^{-3} + \frac{1}{12} \cdot 10^{-3} \\
&= 7.5 \cdot 10^{-4}
\end{aligned}
$$

**Ответ:**  

$$\boxed{
\Delta y \approx 7.5 \cdot 10^{-4}.
}$$

## 2

Разложим каждый член числителя N в ряд Тейлора вокруг точки x до членов порядка $h^5$:

Для $u(x-2h)$:
$$
u(x-2h) = u(x) - 2h u'(x) + \frac{(2h)^2}{2!}u''(x) - \frac{(2h)^3}{3!}u'''(x) + \frac{(2h)^4}{4!}u^{(4)}(x) - \frac{(2h)^5}{5!}u^{(5)}(x) + O(h^6)
$$

Для $u(x-h)$:
$$
u(x-h) = u(x) - h u'(x) + \frac{h^2}{2!}u''(x) - \frac{h^3}{3!}u'''(x) + \frac{h^4}{4!}u^{(4)}(x) - \frac{h^5}{5!}u^{(5)}(x) + O(h^6)
$$

Для $u(x+h)$:
$$
u(x+h) = u(x) + h u'(x) + \frac{h^2}{2!}u''(x) + \frac{h^3}{3!}u'''(x) + \frac{h^4}{4!}u^{(4)}(x) + \frac{h^5}{5!}u^{(5)}(x) + O(h^6)
$$

Для $u(x+2h)$:
$$
u(x+2h) = u(x) + 2h u'(x) + \frac{(2h)^2}{2!}u''(x) + \frac{(2h)^3}{3!}u'''(x) + \frac{(2h)^4}{4!}u^{(4)}(x) + \frac{(2h)^5}{5!}u^{(5)}(x) + O(h^6)
$$

Теперь подставим все разложения в выражение:
$$
N = u(x-2h) - 8u(x-h) + 8u(x+h) - u(x+2h)
$$

Сгруппируем коэффициенты при одинаковых производных:

При $u(x)$:
$$1 - 8 + 8 - 1 = 0$$

При $u'(x)$:
$$-2h - 8(-h) + 8(h) - (2h) = -2h + 8h + 8h - 2h = 12h$$

При $u''(x)$:
$$\frac{(2h)^2}{2!} - 8\cdot\frac{h^2}{2!} + 8\cdot\frac{h^2}{2!} - \frac{(2h)^2}{2!} = 2h^2 - 4h^2 + 4h^2 - 2h^2 = 0$$

При $u'''(x)$:
$$-\frac{(2h)^3}{3!} - 8(-\frac{h^3}{3!}) + 8(\frac{h^3}{3!}) - \frac{(2h)^3}{3!} = -\frac{8h^3}{6} + \frac{8h^3}{6} + \frac{8h^3}{6} - \frac{8h^3}{6} = 0$$

При $u^{(4)}(x):$
$$\frac{(2h)^4}{4!} - 8\cdot\frac{h^4}{4!} + 8\cdot\frac{h^4}{4!} - \frac{(2h)^4}{4!} = \frac{16h^4}{24} - \frac{8h^4}{24} + \frac{8h^4}{24} - \frac{16h^4}{24} = 0$$

После сокращения всех членов до $h^4$ получаем:
$$
N = 12h \cdot u'(x) + O(h^5)
$$

Главный член погрешности имеет порядок h^{5}.

Таким образом, полная формула для производной:

$$
u'(x) = \frac{N}{12h} = u'(x) + O(h^4)
$$
Следовательно, формула имеет **четвертый порядок аппроксимации** $ O(h^4) $.

Теперь найдем оптимальный шаг дифференцирования $ h_{\text{opt}} $

1. Погрешность аппроксимации

Формула имеет порядок аппроксимации $ O(h^4)$. Значит:
$$
\varepsilon_{\text{approx}} = \left| \frac{h^4}{30} u^{(5)}(\xi) \right| \leq \frac{M_5 h^4}{30}
$$

2. Погрешность округления

Значения функции вычисляются с погрешностью $ \Delta u $. Оценим погрешность числителя:

$$
\Delta N = \Delta u + 8\Delta u + 8\Delta u + \Delta u = 18\Delta u
$$

Погрешность округления для производной:
$$
\varepsilon_{\text{round}} = \left| \frac{\Delta N}{12h} \right| = \frac{18\Delta u}{12h} = \frac{3\Delta u}{2h}
$$

3. Полная погрешность

$$
\varepsilon_{total}(h) = \varepsilon_{\text{approx}} + \varepsilon_{\text{round}} = \frac{M_5 h^4}{30} + \frac{3\Delta u}{2h}
$$

Минимизируем $\varepsilon_{total}(h)$ по $h$:

$$
\frac{d\varepsilon_{total}}{dh} = \frac{4M_5 h^3}{30} - \frac{3\Delta u}{2h^2} = 0
$$

$$
\frac{4M_5 h^5}{30} = \frac{3\Delta u}{2}
$$

$$
\frac{2M_5 h^5}{15} = \frac{3\Delta u}{2}
$$

$$
h^5 = \frac{3\Delta u}{2} \cdot \frac{15}{2M_5} = \frac{45\Delta u}{4M_5}
$$

$$
h^* = \left( \frac{45\Delta u}{4M_5} \right)^{1/5}
$$

**Ответ:**  

$$\boxed{
\text{Оптимальный шаг: } h^* = \left( \frac{45\Delta u}{4M_5} \right)^{1/5}
}$$

   $$
   \Delta x_2 \leq 0.5 \cdot 10^{-7 - 4} = 0.5 \cdot 10^{-11}.
   $$

   $$
   \frac{dx_2}{dc} = \frac{1}{\sqrt{729 - 4c}}.

   $$
   \frac{dx_2}{dc} \approx \frac{1}{26.925} \approx 0.03714.
   $$  
   Таким образом:  
   $$
   \Delta x_2 \approx \left| \frac{dx_2}{dc} \right| \Delta c \approx 0.03714 \cdot \Delta c.

   $$
   0.03714 \cdot \Delta c \leq 0.5 \cdot 10^{-11},

   $$

# Упражнение №5

Перепишем соотношение:
$$
5x_{n+1} = x_n + 4 \quad \Rightarrow \quad x_{n+1} = \frac{1}{5}x_n + \frac{4}{5}
$$

Решение реккурентного соотношения имеет вид:
$$
x_n = x_{n}^{(одн)} + x_{n}^{(частн)}
$$

Однородное решение:
$$
x_{n}^{(одн)} = A \cdot \left( \frac{1}{5} \right)^n
$$

Ищем частное решение в виде константы (так как правая часть — константа):
$$ x_n^{(частн)} = C = const $$

Подставляем в исходное:
$$ x_{n+1}^{(частн)} = \frac{1}{5} \cdot x_n^{(частн)} + \frac{4}{5} $$
$$ C = \frac{1}{5} \cdot C + \frac{4}{5} $$
$$ C = \frac{\frac{4}{5}}{1 - \frac{1}{5}} = 1 $$

Таким образом, частное решение:
$$ x_n^{(частн)} = 1 $$

Используем начальное условие $ x_0 $:
$$
x_0 = A \cdot 1 + 1 \quad \Rightarrow \quad A = x_0 - 1
$$

Таким образом, общее решение имеет вид:
$$
x_n = (x_0 - 1) \cdot \left( \frac{1}{5} \right)^n + 1
$$

**Анализ погрешности**

Абсолютная погрешность $x_{0}$:
$$
\Delta_0 = 10^{-6} |x_0|
$$

На каждой итерации вычисляется:
$$
\tilde{x}_{n+1} = \frac{1}{5} \tilde{x}_n + \frac{4}{5}
$$

Поскольку операция умножения на $ \frac{1}{5} $ и сложение с $ \frac{4}{5} $ выполняются с ошибками округления, но мы будем считать, что все операции выполняются точно, машинная погрешность пренебрижимо мала, и погрешность обусловлена только ошибкой в $ x_0 $.

Тогда абсолютная погрешность $ \Delta_n $ оценивается:
$$
\Delta_{n+1} = \left| \frac{1}{5} \right| \Delta_n = \frac{1}{5} \Delta_n
$$

Таким образом:
$$
\Delta_n = \left( \frac{1}{5} \right)^n \Delta_0 = \left( \frac{1}{5} \right)^n \cdot 10^{-6} |x_0|
$$

**Относительная погрешность**

Оценим относительную погрешность:
$$
\delta_n = \frac{\Delta_n}{|x_n|} = \frac{ \left( \frac{1}{5} \right)^n \cdot 10^{-6} |x_0| }{ \left| (x_0 - 1) \cdot \left( \frac{1}{5} \right)^n + 1 \right| }
$$

Упростим:
$$
\delta_n = \frac{10^{-6} |x_0| }{ \left| (x_0 - 1) + 5^n \right| }
$$

**Анализ погрешности**


Относительная погрешность становится большой, если знаменатель мал:
$$
|(x_0 - 1) + 5^n| \approx 0
$$
Это выполняется, когда:
$$
x_0 - 1 \approx -5^n \quad \Rightarrow \quad x_0 \approx 1 - 5^n
$$

Эти значения не достигаются при 

**Ответ**
$$
\boxed{
x_0 \approx 1- 5^n, \quad n \in \mathbb{N} }
$$
