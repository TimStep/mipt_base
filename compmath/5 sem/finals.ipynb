```python
import numpy as np
from numpy.linalg import norm

seed = 42
np.random.seed(seed)
```

```python
def power_method(A, eps=1e-10, max_iter=1000):
    n = A.shape[0]
    x = np.random.rand(n)
    x = x / norm(x)
    
    for k in range(max_iter):
        y = A @ x
        lambda_curr = np.dot(x, y)
        x_new = y / norm(y)
        
        error = norm(A @ x_new - lambda_curr * x_new)
        
        if error <= eps:
            return lambda_curr, x_new, k + 1
        
        x = x_new
    
    return lambda_curr, x_new, max_iter
```

```python
A = np.array([[8, 2, -1],
              [2, -5, -8],
              [-1, -8, -5]], dtype=float)

eps = 0.00001
lambda_max, eigenvec, iterations = power_method(A, eps=eps)

print(f"Радиус: {lambda_max:.10f}")
print(f"Итерации: {iterations}")
print(f"Собственный вектор: {eigenvec}")

eigenvals_exact = np.linalg.eigvals(A)
lambda_max_exact = np.max(eigenvals_exact.real)
print(f"\n np.linalg говорит: {lambda_max_exact:.10f}")
```
