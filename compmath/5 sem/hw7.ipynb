# 4 (пункт d)

```python
x = sp.symbols('x')
f_sym = x**2 - 20*sp.sin(x)
f = sp.lambdify(x, f_sym, 'numpy')
f_prime = sp.lambdify(x, sp.diff(f_sym, x), 'numpy')
```

```python
def dichotomy(f, a, b, eps):

    while abs(b - a) / 2 > eps:
        c = (a + b) / 2
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
            
    return (a + b) / 2

def newton(f, df, x0, eps):
    x = x0
    while True:
        fx = f(x)
        dfx = df(x)
           
        x_next = x - fx / dfx
        
        if abs(x_next - x) < eps:
            return x_next
        
        x = x_next

```

```python
a, b = 2.5, 2.8
x0 = (a+b)/2
eps = 1e-4
print(f"Метод дихотомии: {dichotomy(f, a, b, eps):.6f}")
print(f"Метод Ньютона: {newton(f, f_prime, x0, eps):.6f}")
```

```python
def simple_iteration_method(X0, eps, phi_func):
    X_prev = X0
 
    while True:
        X_next = phi_func(X_prev)

        if np.linalg.norm(X_next - X_prev, ord=np.inf) < eps:
            return X_next
            
        X_prev = X_next
        
    return X_prev

def newton_system_method(X0, eps, F_func, Jac_func):
    X_prev = X0
    
    while True:
        F_vec = F_func(X_prev)
        J_mat = Jac_func(X_prev)

        dX = np.linalg.solve(J_mat, -F_vec)
        X_next = X_prev + dX

        if np.linalg.norm(dX, ord=np.inf) < eps:
            return X_next

        X_prev = X_next
        
    return X_prev
```

## Система В

```python
epsilon = 1e-3
X0 = np.array([3.0, 0.0]) 


def F(X):
    x, y = X
    F1 = math.cos(x - 1) + y + 0.5
    F2 = x - math.cos(y) - 3
    return np.array([F1, F2])

def Phi(X):
    x, y = X
    phi1 = math.cos(y) + 3
    phi2 = -math.cos(x - 1) - 0.5
    return np.array([phi1, phi2])

#ручками быстрее, чем sympy
def Jacobian(X):
    x, y = X
    J = np.array([
        [-math.sin(x - 1), 1.0],
        [1.0, math.sin(y)]
    ])
    return J


X_m_root = simple_iteration_method(X0, epsilon, Phi)
print(f"Метод простых итераций x = {X_m_root[0]:.3f}, y = {X_m_root[1]:.3f}")

X_n_root = newton_system_method(X0, epsilon, F, Jacobian)
print(f"Метод Ньютона: x = {X_n_root[0]:.3f}, y = {X_n_root[1]:.3f}")
```

## Система Г

```python
epsilon = 1e-3
X0 = np.array([0.5, -1.0]) 


def F(X):
    x, y = X
    F1 = math.sin(x + 2) - y - 1.5
    F2 = x + math.cos(y - 2) - 0.5
    return np.array([F1, F2])

def Phi(X):
    x, y = X
    phi1 = 0.5 - math.cos(y - 2)
    phi2 = math.sin(x + 2) - 1.5
    return np.array([phi1, phi2])

def Jacobian(X):
    x, y = X
    J = np.array([
        [math.cos(x + 2), -1.0],
        [1.0, -math.sin(y - 2)]
    ])
    return J


X_m_root = simple_iteration_method(X0, epsilon, Phi)
print(f"Метод простых итераций x = {X_m_root[0]:.3f}, y = {X_m_root[1]:.3f}")

X_n_root = newton_system_method(X0, epsilon, F, Jacobian)
print(f"Метод Ньютона x = {X_n_root[0]:.3f}, y = {X_n_root[1]:.3f}")
```
